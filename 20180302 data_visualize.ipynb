```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

xs = df7['Year']
ys = df7['Rank']
zs = df7['ID1']
ax.scatter(xs, ys, zs, s=50, alpha=0.6, edgecolors='w')

ax.set_xlabel('Year')
ax.set_ylabel('Rank')
ax.set_zlabel('Count')
```

```python
fig = plt.figure(figsize=(12,8))
ax = fig.add_subplot(111, projection='3d')

xs = df7['Year']
ys = df7['Rank']
zs = df7['ID1']
p = ax.scatter(xs, ys, zs, s=zs*3, alpha=0.6, edgecolors='w', c=zs, cmap='jet')
plt.colorbar(p)

ax.set_xlabel('Year')
ax.set_ylabel('Rank')
ax.set_zlabel('Freq')
```

```python
import pandas as pd
url = "https://zh.wikipedia.org/wiki/%E4%B8%AD%E5%8D%8E%E4%BA%BA%E6%B0%91%E5%85%B1%E5%92%8C%E5%9B%BD%E7%9C%81%E7%BA%A7%E8%A1%8C%E6%94%BF%E5%8C%BA%E5%9C%B0%E5%8C%BA%E7%94%9F%E4%BA%A7%E6%80%BB%E5%80%BC%E5%88%97%E8%A1%A8"
table = pd.read_html(url)[0]
table = table.drop(table.columns[[0,1,2]],axis=0)
table.head()
#table.to_excel('2017chinagdp.xlsx')
```

```python
ChinaGdp = pd.read_excel('2017chinagdp.xlsx')
ChinaGdp.head()
```

```python
china_map= china_map.rename(index = str, columns = {'NAME_1':'District'})#修改列名
```

```python
data_geod = gpd.GeoDataFrame(ChinaGdp)##将data转换为geopandas.DataFrame
da_merge = china_map.merge(data_geod, on = data_geod.iloc[:,6], how='left')#数据合并
```

```python
da_merge.head()
```

```python
#da_merge.plot(da_merge.iloc[:,15], cmap = plt.cm.Greens)
da_merge.plot('gdp',figsize=(20,10), cmap = 'Reds', legend=True)
plt.title("2017 China GDP (units: 10 billion CNY)", fontsize=30)
plt.savefig("2017ChinaGDP_Red.png",figsize=(30,30))
```

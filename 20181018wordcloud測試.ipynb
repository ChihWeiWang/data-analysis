```python
#使用codecs.open指定編碼格式，讀取中文
mytext = codecs.open(u'K:\\20181018二三四冊\\wordcloud\\二三四冊事件敘述1368-1911.txt',encoding='utf-8').read()
```

```python
# 中文字體參考路徑
font = r'C:\Windows\Fonts\kaiu.ttf'
wordcloud = WordCloud(background_color="white",width=2000, height=1000, max_words = 200,font_path=font).generate(mytext)
```

```python
# 設定停用字(排除常用詞、無法代表特殊意義的字詞)
stopwords = {}.fromkeys([u"六月",u"七月"])
# 中文字體參考路徑
font = r'C:\Windows\Fonts\kaiu.ttf'
wordcloud = WordCloud(background_color="white",width=1000, height=860, max_words = 200,font_path=font,stopwords=stopwords).generate(mytext)
```

```python
plt.figure(figsize=(20,20), dpi = 100)
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
```

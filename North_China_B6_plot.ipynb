```python
#-*- coding: utf-8 -*-
import pandas as pd
import sys
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy import stats, integrate
```

```python
%matplotlib inline
domain = 'N_B6'
DataPath = str('K:\\20171027資料前處理\\降水分級20171108\\13-全年_分區_降水分級_不含條目之西曆月份起迄-9999\華北次分區\\'+ (domain) + '\\')
NDataPath = unicode(DataPath,'utf8')
```

```python
#使用pandas模組讀取excel資料;u表示unicode編碼;sheetname=1表示第二張工作表，0表示第一張;skiprows=3表示前四行略過
df = pd.read_excel(unicode(DataPath,'utf8')  + (domain) + u'降水分級.xlsx')
#顯示前5行資料
df.head()
```

```python
nl = len(df)
```

```python
Fig_ts = plt.figure(figsize=(12,12))
sns.stripplot(x="Index_P", y="Year", data=df , jitter=True);
plt.grid(True)
plt.tick_params(axis='x', which='major', labelsize=20)
plt.tick_params(axis='y', which='major', labelsize=20)
plt.title("Frequency " + (domain) +" ( n="+str(nl)+" )", fontsize=20)
plt.savefig( unicode(DataPath,'utf8') + '2Dstripplot_' + (domain) + '.jpg')
```

```python
Fig_ts = plt.figure(figsize=(12,12))
sns.swarmplot(x="Index_P", y="Year", data=df)
plt.grid(True)
plt.tick_params(axis='x', which='major', labelsize=20)
plt.tick_params(axis='y', which='major', labelsize=20)
plt.title("Frequency " + (domain) +" ( n="+str(nl)+" )", fontsize=20)
plt.savefig(unicode(DataPath,'utf8') + 'swarmplot_' + (domain) + '.jpg')
```

```python
Fig_ts = plt.figure(figsize=(20,12))
sns.violinplot(x="Index_P", y="Year", data=df)
plt.grid(True)
plt.tick_params(axis='x', which='major', labelsize=20)
plt.tick_params(axis='y', which='major', labelsize=20)
plt.xlabel("Precipitation Index", fontsize=20)
plt.ylabel("Year", fontsize=20)
plt.title("Frequency " + (domain) +" ( n="+str(nl)+" )", fontsize=20)
plt.savefig(unicode(DataPath,'utf8')+'violinplot_' + (domain) + '.jpg')
```

```python
Fig_ts = plt.figure(figsize=(20,12))

sns.violinplot(x="Index_P", y="Year", data=df, inner=None)
#sns.swarmplot(x="Index_P", y="Year", data=df, color="w")
sns.swarmplot(x="Index_P", y="Year", data=df,color='b')
plt.grid(True)
plt.tick_params(axis='x', which='major', labelsize=20)
plt.tick_params(axis='y', which='major', labelsize=20)
plt.xlabel("Precipitation Index", fontsize=20)
plt.ylabel("Year", fontsize=20)
plt.title("Frequency " + (domain) +" (n="+str(nl)+" )", fontsize=20)
plt.savefig(unicode(DataPath,'utf8')+'violinswarmplot_' + (domain) + '.jpg')
```

```python
Fig_ts = plt.figure(figsize=(20,12))
sns.boxplot(x="Index_P", y="Year", data=df)
sns.swarmplot(x="Index_P", y="Year", data=df)
plt.grid(True)
plt.tick_params(axis='x', which='major', labelsize=20)
plt.tick_params(axis='y', which='major', labelsize=20)
plt.xlabel("Precipitation Index", fontsize=20)
plt.ylabel("Year", fontsize=20)
plt.title("Frequency " + (domain) +" ( n="+str(nl)+" )", fontsize=20)
plt.savefig(unicode(DataPath,'utf8')+'boxswarmplot_' + (domain) + '.jpg')
```

```python
Fig_ts = plt.figure(figsize=(20,4))
sns.violinplot(df.iloc[0:200,3], df.iloc[0:200,2], data=df)
plt.grid(True)
plt.tick_params(axis='x', which='major', labelsize=10)
plt.tick_params(axis='y', which='major', labelsize=20)
plt.ylabel("Precipitation Index", fontsize=20)
plt.xlabel("Year", fontsize=20)
plt.title("Frequency " + (domain) +"("+str(nl)+")", fontsize=20)
plt.savefig(unicode(DataPath,'utf8')+'violinplot2_' + (domain) + '.jpg')
```

```python
Fig_ts = plt.figure(figsize=(20,12))
sns.kdeplot(df.iloc[:,2], df.iloc[:,3], shade="True", n_levels=20, cmap="Reds", cbar=True,kernel='gaussian', bandwidth=0.75)
plt.tick_params(axis='x', which='major', labelsize=20)
plt.tick_params(axis='y', which='major', labelsize=20)
plt.title("Frequency " + (domain) +"("+str(nl)+")", fontsize=20)
plt.savefig(unicode(DataPath,'utf8')+'kdeplot_xy_' + (domain) + '.jpg')
```
